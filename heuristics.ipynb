```
import numpy as np
import pandas as pd
```

```
distances=pd.read_csv("roaddistance1.csv")
distances
```

```
list_columns=distances.columns.tolist()
```

```
list_columns[1]
```

```
import networkx as nx
G=nx.Graph()
for i in range(len(distances)):
  for j in range(1,len(list_columns)):
    G.add_edge(distances.iloc[i,0],list_columns[j],weight=distances.iloc[i,j])
```

```
len_path = dict((nx.all_pairs_dijkstra(G)))
```

```
pd.DataFrame(len_path['Nasik'])
```

```
print(nx.dijkstra_path_length(G, 'Nasik', 'Bombay'))
```

```
heuristic_data=pd.read_csv('roaddistance1 - Copy.csv')
```

```
heuristic_data
```

```
hue_columns=heuristic_data.columns.tolist()
```

```
hue_columns
```

```
for i in range(len(heuristic_data)):
  for j in range(1,len(hue_columns)):
    heuristic_data.iloc[i,j]=nx.dijkstra_path_length(G, heuristic_data.iloc[i,0],hue_columns[j])
```

```
heuristic_data.to_csv('heuristic_data.csv') 
```

```
import matplotlib.pyplot as plt


# edges
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 2000]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 2000]

pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos,node_size=2000)
nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    G, pos, edgelist=esmall, width=10, alpha=1.0, edge_color="b", style="dashed"
)

# labels
nx.draw_networkx_labels(G, pos, font_size=5, font_family="sans-serif")

plt.axis("off")
plt.show()
plt.savefig('network.png')
```
